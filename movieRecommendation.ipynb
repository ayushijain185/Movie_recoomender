#<font color="green"><h1>Analysis of columns for context based recommendation</h1>

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
#popularity - but its make project complex because of numeric data
#release date - its also complex because of numeric data

movies = movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l


def convert3(obj):
    l=[]
    count=0
    for i in ast.literal_eval(obj):
        if(count!=3):
            l.append(i['name'])
            count+=1
    return l

def fetchDirector(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies["crew"]=movies["crew"].apply(fetchDirector)
```

```python
movies.head()
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["keywords"]=movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tag"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
final_df=movies[["movie_id","title","tag"]]
```

```python
final_df
```

```python
final_df["tag"]=final_df["tag"].apply(lambda x:" ".join(x))
```

```python
final_df.head()
```

```python
final_df['tag']=final_df['tag'].apply(lambda x:x.lower())
```

```python
final_df['tag'][0]
```

```python
final_df.head()
```

<font color="green"><h1>vectorization for recommendation using <b>bag of words</b> technique</h1>

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(final_df['tag']).toarray()
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    index = final_df[final_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(final_df.iloc[i[0]].title)
        
```

```python
recommend('Gandhi')
```
